```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
vector1 = np.random.rand(15)
vector2 = np.random.rand(15)
```

```python
plt.figure(figsize=(8, 5))
plt.plot(vector1, label="Vector 1", marker='o')
plt.plot(vector2, label="Vector 2", marker='s')
plt.xlabel("Index")
plt.ylabel("Value")
plt.title("Comparison of Two Random Vectors")
plt.legend()
plt.grid(True)
plt.show()
```
