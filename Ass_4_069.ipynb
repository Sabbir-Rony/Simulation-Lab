```python
import random
import numpy as np
import math
```

```python
R = 69
radius = R / 10
min_points = 2000

def is_inside_circle(x, y, r):
    return x*x + y*y <= r*r

def run_simulation(total_points):
    inside = 0

    half = total_points // 2

    for _ in range(half):
        x = random.uniform(0, radius)
        y = random.uniform(0, radius)
        if is_inside_circle(x, y, radius):
            inside += 1

    xs = np.random.rand(total_points - half) * radius
    ys = np.random.rand(total_points - half) * radius

    for x, y in zip(xs, ys):
        if is_inside_circle(x, y, radius):
            inside += 1

    pi_estimate = (inside / total_points) * 4
    error = abs(math.pi - pi_estimate)

    return inside, pi_estimate, error
```

```python
print("Range Used | Total Points | Points Inside | Estimated π | Error")
print("--------------------------------------------------------------------")

for i in range(1, 6):
    total = min_points * i
    inside, pi_estimate, error = run_simulation(total)

    print(f"[0, {radius}] | {total} | {inside} | {pi_estimate:.6f} | {error:.6f}")
```
