Part - 1

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
uniform_wait = np.random.uniform(2, 12, 3000)
normal_wait = np.random.normal(8, 2, 3000)
```

```python
normal_wait = np.where(normal_wait < 0, 0, normal_wait)
```

Part - 2

```python
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.hist(uniform_wait, bins=30, edgecolor='black')
plt.title("Regular Day (Uniform Distribution)")
plt.xlabel("Wait Time (minutes)")
plt.ylabel("Frequency")
```

```python
plt.subplot(1,2,2)
plt.hist(normal_wait, bins=30, edgecolor='black')
plt.title("Busy Day (Normal Distribution)")
plt.xlabel("Wait Time (minutes)")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

Part - 3

```python
def calculate_stats(data, name):
    print(f"\n{name} Statistics:")
    print("Mean:", np.mean(data))
    print("Median:", np.median(data))
    print("Standard Deviation:", np.std(data))
    print("Maximum Wait Time:", np.max(data))

calculate_stats(uniform_wait, "Regular Day (Uniform)")
calculate_stats(normal_wait, "Busy Day (Normal)")
```
